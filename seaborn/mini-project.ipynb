# Project: Nutrition Data Visualisation on 80 cereal products

Link: https://www.kaggle.com/crawford/80-cereals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset cereal từ file CSV
cereal = pd.read_csv("./data/cereal.csv")
```

```python
# Xem dữ liệu cereal
cereal.head()
```

```python
# Loại bỏ các cột không cần thiết: 'shelf', 'weight', 'cups', 'rating'
fields = ["shelf", "weight", "cups", "rating"]
cereal_df = cereal.drop(fields, axis=1)
cereal_df.head()
```

```python
# Tính ma trận correlation
cereal_corr = cereal_df.select_dtypes(include=[np.number]).corr()
cereal_corr
```

```python
# Tạo ma trận boolean cùng shape với correlation matrix
# ones_like() tạo ma trận toàn True với cùng kích thước
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
# Kiểm tra shape của hai ma trận
ones_corr.shape, cereal_corr.shape
```

```python
# np.triu: trả về ma trận tam giác trên (upper triangle)
# Dùng để mask nửa trên của heatmap (tránh hiển thị trùng lặp)
mask = np.triu(ones_corr)
mask
```

```python
# Vẽ heatmap với mask để ẩn tam giác trên
sns.heatmap(data=cereal_corr, mask=mask);
```

```python
# When removing the upper tri-angle, here are still two empty cells in our matrix (calories & vitamin)
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
# Điều chỉnh mask: bỏ dòng đầu và cột cuối để loại ô trống
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
```

```python
# Điều chỉnh correlation matrix: bỏ dòng đầu và cột cuối
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask);
```

```python
# Tạo heatmap chuyên nghiệp với:
# - Figure size tùy chỉnh
# - Diverging color palette (HUSL colors)
# - Annotations với giá trị correlation
# - Custom tick labels
# - Title
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(
    data=adjusted_cereal_corr,
    mask=adjusted_mask,
    annot=True,
    annot_kws={"fontsize": 13},
    fmt=".2f",
    cmap=cmap,
    vmin=-1,
    vmax=1,
    linecolor="white",
    linewidths=0.5,
)
yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0)
ax.set_xticklabels(xticks, rotation=45)
title = "Cereal Nutrition Correlation Matrix\n"
ax.set_title(title, loc="left", fontsize=13);
```
